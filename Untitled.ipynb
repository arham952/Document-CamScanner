```python
#Remove the Noise
## Image Blurring
orig = img.copy()
gray = cv2.cvtColor(orig, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap = "binary")
plt.show()

blurred = cv2.GaussianBlur(gray,(5, 5), 0)
plt.imshow(blurred, cmap = "binary")
plt.show()
```

```python
regen = cv2.cvtColor(blurred, cv2.COLOR_GRAY2BGR)
plt.imshow(orig)
plt.show()

plt.imshow(regen)
plt.show()
```

```python
edge = cv2.Canny(blurred, 0, 50)  # Now 'edge' is defined
print(edge.shape)  # Now this will work
```

```python
# Draw all contours on the original image to see what is detected
contour_img = orig.copy()
cv2.drawContours(contour_img, contours, -1, (0, 255, 0), 2)

plt.imshow(cv2.cvtColor(contour_img, cv2.COLOR_BGR2RGB))
plt.title("All Detected Contours")
plt.show()
```

```python
## Countours Extraction

contours, _= cv2.findContours(edge, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
print(len(contours))

contours = sorted(contours ,reverse =True ,key = cv2.contourArea)
```

```python
## Select the best contour Region

target = None  # Initialize target to avoid "undefined variable" error
for c in contours:
    p = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.02 * p, True)  # Increase epsilon if needed

    print(f"Contour Points: {len(approx)}")  # Debugging output

    if len(approx) == 4:
        target = approx
        break

if target is None:
    print("No suitable contour found!")
else:
    print("Target shape:", target.shape)
```

```python
## Reorder target contour

def reorder(points):
    points = points.reshape((4, 2))
    new_points = np.zeros((4, 2), dtype=np.float32)

    add = points.sum(axis=1)
    diff = np.diff(points, axis=1)

    new_points[0] = points[np.argmin(add)]  # Top-left
    new_points[1] = points[np.argmin(diff)]  # Top-right
    new_points[2] = points[np.argmax(add)]  # Bottom-right
    new_points[3] = points[np.argmax(diff)]  # Bottom-left

    return new_points


```

```python
reorder = reorder(target)
print("*****************")
print(reorder)
```

```python
## Project to a fixed screen

input_representation = reorder

output_map = np.float32([[0,0],[800,0],[800,800],[0,800]])
```

```python
input_representation = reorder  # Ensure correct order
output_map = np.float32([[0, 0], [800, 0], [800, 800], [0, 800]])

M = cv2.getPerspectiveTransform(input_representation, output_map)
scanned_img = cv2.warpPerspective(orig, M, (800, 800))

# Display the final output
plt.imshow(cv2.cvtColor(scanned_img, cv2.COLOR_BGR2RGB))
plt.title("Scanned Document")
plt.show()
```
