```python
## import basic libraryies
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
im_path = "./Bill2.jpg"
#read image from path
img =  cv2.imread(im_path)
print(img.shape)

#img resize
img = cv2.resize(img,(1500,800))

#BGR format => RBG format
print(img.shape)

plt.imshow(img)
plt.show()
```

* Remove the Noise
* Edge Detection
* Contour Extraction
* Best Contour Selection
* Project to the Screen

```python
#Remove the Noise
## Image Blurring
orig = img.copy()
gray = cv2.cvtColor(orig, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap = "binary")
plt.show()

blurred = cv2.GaussianBlur(gray,(5, 5), 0)
plt.imshow(blurred, cmap = "binary")
plt.show()
```

```python
  regen = cv2.cvtColor(blurred, cv2.COLOR_GRAY2BGR)
plt.imshow(orig)
plt.show()

plt.imshow(regen)
plt.show()
```
